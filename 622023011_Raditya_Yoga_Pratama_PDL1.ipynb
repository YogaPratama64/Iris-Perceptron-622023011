```python
def min_max_normalize(data):
  min_vals = data.iloc[:, :4].min(axis=0)
  max_vals = data.iloc[:, :4].max(axis=0)
  normalized = (data.iloc[:, :4] - min_vals) / (max_vals - min_vals)
  return normalized

norm_setosa = min_max_normalize(setosa)
norm_versicolor = min_max_normalize(versicolor)
norm_virginica = min_max_normalize(virginica)
```

```python
combined_data = np.vstack([
    np.hstack([norm_setosa, np.full((norm_setosa.shape[0], 1), 0)]),
    np.hstack([norm_versicolor, np.full((norm_versicolor.shape[0], 1), 1)]),
    np.hstack([norm_virginica, np.full((norm_virginica.shape[0], 1), 2)])
])
```

```python
raditya = pd.DataFrame(combined_data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
```
